## Feature Engineering

```python
from pyspark.ml.feature import StringIndexer
from pyspark.ml.feature import OneHotEncoder
from pyspark.ml.feature import VectorAssembler
from pyspark.ml import Pipeline
from pyspark.ml.feature import StandardScaler
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
# Label Indexing

indexer = StringIndexer(inputCol='class', outputCol='label')
df_indexed = indexer.fit(df_stars).transform(df_stars)
```

```python
df_indexed.columns
```

```python
# Create feature vector
assembler = VectorAssembler(inputCols=['alpha', 'delta', 'u', 'g', 'r', 'i', 'z', 'redshift','plate', 'MJD', 'cam_col'], outputCol='features')
df_vectorized = assembler.transform(df_indexed)
```

```python
# Scale features
scaler = StandardScaler(inputCol='features', outputCol='scaled_features', withMean=True, withStd=True)
df_scaled = scaler.fit(df_vectorized).transform(df_vectorized)
```

```python
df_scaled.show()
```

```python
df_scaled.show(2)
```

```python
df_scaled.select('label').show(10)
```

```python
df_scaled.columns
```

```python
df_scaled.select('class').show(5)
```

```python
df_labels = df_scaled.select('label')
df_class = df_scaled.select('class')
df_labels.show(2),df_class.show(2)
```

```python
df_train, df_test = df_scaled.randomSplit([0.7, 0.3], seed=42)
```

# Model Training and Testing

### **Logistic Regression**

```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

logistic_regressor = LogisticRegression()
logistic_regressor_model = logistic_regressor.fit(df_train)
logistic_regressor_pred = logistic_regressor_model.transform(df_test)

evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
lr_accuracy = evaluator.evaluate(logistic_regressor_pred)
print("Logistic Regression Accuracy = " + str(lr_accuracy))
```

### **Decision Tree**

```python
from pyspark.ml.classification import DecisionTreeClassifier

decision_classifier = DecisionTreeClassifier(featuresCol='features', labelCol='label', maxDepth=5)
decision_model = decision_classifier.fit(df_train)
decision_predictions = decision_model.transform(df_test)

evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
decision_accuracy = evaluator.evaluate(decision_predictions)

print("Decision Tree Accuracy: " + str(decision_accuracy))
```

### **Random Forest**

```python
from pyspark.ml.classification import RandomForestClassifier

random_classifier = RandomForestClassifier(featuresCol='features', labelCol='label', numTrees=200)
random_model = random_classifier.fit(df_train)
random_predictions = random_model.transform(df_test)

evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
random_accuracy = evaluator.evaluate(random_predictions)

print("Random Forest Accuracy: " + str(random_accuracy))
```

# Model Selection

```python
lr_metrics = MulticlassMetrics(rdd_logistic_regressor_pred_label)
dt_metrics = MulticlassMetrics(rdd_decision_predictions_label)
rf_metrics = MulticlassMetrics(rdd_random_predictions_label)
```

```python
selected_accuracy = max([lr_metrics.accuracy, dt_metrics.accuracy, rf_metrics.accuracy ])
selected_accuracy_index = [lr_metrics.accuracy, dt_metrics.accuracy, rf_metrics.accuracy ].index(selected_accuracy)
print(selected_accuracy_index)
```

```python
selected_precision = max([lr_metrics.precision(0), dt_metrics.precision(0), rf_metrics.precision(0) ])
selected_precision_index = [lr_metrics.precision(0), dt_metrics.precision(0), rf_metrics.precision(0)].index(selected_precision)
print(selected_precision_index)
```

```python
selected_recall = max([lr_metrics.recall(0), dt_metrics.recall(0), rf_metrics.recall(0)])
selected_recall_index = [lr_metrics.recall(0), dt_metrics.recall(0), rf_metrics.recall(0)].index(selected_recall)
print(selected_recall_index)
```

Since Random Forest has the highest recall and precision, f1-score evaluation is required.

` F1 Score = 2 * (Precision * Recall) / (Precision + Recall).`

```python
f1_scores = []
for x in [lr_metrics, dt_metrics, rf_metrics]:
    f1 = 2 * (x.precision(0) * x.recall(0)) / (x.precision(0) + x.recall(0)) 
    f1_scores.append(f1)
f1_scores.index(max(f1_scores))
```

### Since `Random Forest Classifier` has the greatest accuracy and F1 Score among the three classifiers, it would be ideal to select the `Random Forest Classifier`
